```python
import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
XLIM = (-4, 4)
YLIM = (-4, 4)


def f(x, y):
    # RHS of y' = f(x,y)
    return x + y


fig = plt.figure()

# draw axes
ax = plt.axes(xlim=XLIM, ylim=YLIM)
ax.set_aspect('equal')

ax.hlines(0, *XLIM, lw=0.5)
ax.vlines(0, *YLIM, lw=0.5)

# linspace^2 for the vf
x = np.linspace(*XLIM, 21)
y = np.linspace(*YLIM, 21)
X, Y = np.meshgrid(x, y)

# norm
norm = np.hypot(1, f(X, Y))

# vector field
kwargs = {'angles':'xy', 'width':0.002, 'pivot':'mid'}
ax.quiver(X, Y, 1/norm, f(X, Y)/norm, **kwargs)

h = 0.1  # Step size
x = np.arange(0, 5, h)  # Numerical grid
y0 = -1  # Initial Condition

# Explicit Euler Method
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])
plt.plot(x, y, 'b--')

plt.show()
```
